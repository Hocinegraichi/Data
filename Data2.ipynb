```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('covid_eu_daily.csv')
df
```

```python
df[['cases', 'deaths']] .groupby([df ['dateRep']]).mean ()
```

```python
df[['cases', 'deaths']] .groupby([df ['countriesAndTerritories']]).mean ()
```

```python
pop =df['popData2020'] .groupby([df ['countriesAndTerritories']]).mean ()

pop.sum()

```

```python
population =pd.DataFrame(df['popData2020'] .groupby([df ['countriesAndTerritories']]).mean ())

population.plot.bar(figsize=(16,10))
```

```python
population =pd.DataFrame(df['cases'] .groupby([df ['countriesAndTerritories']]).mean ())

population.plot.bar(figsize=(16,10))
```

```python
total_cases_per_year =pd.DataFrame(df['cases'] .groupby([df ['year']]).sum ())
total_cases_per_year.plot.bar(figsize=(13,8))
```

```python
average_cases_per_year =pd.DataFrame(df['cases'] .groupby([df ['month']]).mean ())
average_cases_per_year.plot.bar(figsize=(13,8))
```
